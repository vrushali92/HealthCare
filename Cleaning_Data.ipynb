# Cleaning State column

```python
pd.DataFrame(categoricals_df.STATE.value_counts(dropna = False)).reset_index()
```

```python
# Combining the states whose count is less than 2500 in ther category

df = pd.DataFrame(categoricals_df.STATE.value_counts()).reset_index()

df.columns = ['state', 'count']
other_states = list(df[df['count']<2500]['state'])
#other_states
def clean_state(x):
    if x in other_states:
        return 'Other'
    else:
        return x
    
categoricals_df.STATE = list(map(clean_state, categoricals_df.STATE))
```
